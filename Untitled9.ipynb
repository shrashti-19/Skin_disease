```python
## Creating the image paths columns
import pandas as pd
import os

# Load metadata CSV
metadata_path = "/content/skin_cancer_data/HAM10000_metadata.csv"
train_df = pd.read_csv(metadata_path)

# Define image folder paths
image_folder_1 = "/content/skin_cancer_data/HAM10000_images_part_1/"
image_folder_2 = "/content/skin_cancer_data/HAM10000_images_part_2/"

# Function to map image_id to full path
def get_image_path(image_id):
    if os.path.exists(os.path.join(image_folder_1, image_id + ".jpg")):
        return os.path.join(image_folder_1, image_id + ".jpg")
    elif os.path.exists(os.path.join(image_folder_2, image_id + ".jpg")):
        return os.path.join(image_folder_2, image_id + ".jpg")
    else:
        return None  # If not found

# Create the image_path column
train_df["image_path"] = train_df["image_id"].apply(get_image_path)

# Drop missing image paths
train_df = train_df.dropna(subset=["image_path"])

# Verify
print(train_df.head())
print(" Image paths added successfully!")
```

Preprocessing Images : Resizing images to (128,128)

Normalizing pixel values (between 0 and 1)

Convert to Numpy array for training


```python
import cv2
import numpy as np

# Image size
IMG_SIZE = 128

# Function to load and preprocess an image
def load_and_preprocess_image(image_path):
    img = cv2.imread(image_path)  # Read image
    img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)  # Convert to RGB
    img = cv2.resize(img, (IMG_SIZE, IMG_SIZE))  # Resize
    img = img / 255.0  # Normalize (scale pixels between 0-1)
    return img

# Apply preprocessing to all images
X = np.array([load_and_preprocess_image(path) for path in train_df["image_path"]])
y = np.array(train_df["dx"])  # Labels

# Verify shapes
print(" X shape:", X.shape)
print(" y shape:", y.shape)
```

**Converting Labels to Numbers**
Using  Label Encoding to map each disease category to a number.

Converting  labels into one-hot encoded format (useful for classification).

```python
from sklearn.preprocessing import LabelEncoder
from tensorflow.keras.utils import to_categorical

# Encode labels (convert categories to numbers)
label_encoder = LabelEncoder()
y_encoded = label_encoder.fit_transform(y)

# Convert to one-hot encoding
y_onehot = to_categorical(y_encoded)

# Verify
print(" Encoded labels:", y_encoded[:5])
print("One-hot shape:", y_onehot.shape)
print("Class mapping:", dict(zip(label_encoder.classes_, range(len(label_encoder.classes_)))))
```

**SPLIT THE DATA **

```python
from sklearn.model_selection import train_test_split
# Split into training (80%) and testing (20%)
X_train, X_test, y_train, y_test, train_indices, test_indices = train_test_split(
    X, y_onehot, train_df.index, test_size=0.2, random_state=42, stratify=y_onehot
)

# Create test_df using the correct indices
test_df = train_df.loc[test_indices].copy()

# Verify shapes
print(f"X_train shape: {X_train.shape}")
print(f"X_test shape: {X_test.shape}")
print(f"y_train shape: {y_train.shape}")
print(f"y_test shape: {y_test.shape}")
print(f"test_df shape: {test_df.shape}")  # Check test_df shape


```

```python
import gc
import tensorflow as tf

# Delete unnecessary variables
del X, y, X_train, X_test, y_train, y_test

# Force garbage collection
gc.collect()

# Clear TensorFlow session
tf.keras.backend.clear_session()
```

```python
from tensorflow.keras.preprocessing.image import ImageDataGenerator

IMG_SIZE = 128
BATCH_SIZE = 16

# Define data generator (rescale pixels between 0-1)
datagen = ImageDataGenerator(rescale=1.0/255.0, validation_split=0.2)

# Training data generator
train_generator = datagen.flow_from_dataframe(
    dataframe=train_df,
    x_col="image_path",
    y_col="dx",
    target_size=(IMG_SIZE, IMG_SIZE),
    batch_size=BATCH_SIZE,
    class_mode="categorical",
    subset="training",
    shuffle=True  # Shuffle training data
)

# Validation data generator
val_generator = datagen.flow_from_dataframe(
    dataframe=train_df,
    x_col="image_path",
    y_col="dx",
    target_size=(IMG_SIZE, IMG_SIZE),
    batch_size=BATCH_SIZE,
    class_mode="categorical",
    subset="validation",
    shuffle=False  # No need to shuffle validation
)

# Test Data Generator (without validation split)
test_datagen = ImageDataGenerator(rescale=1.0/255.0)

# Ensure `test_df` is defined correctly before using it
if 'test_df' in locals() and not test_df.empty:
    test_generator = test_datagen.flow_from_dataframe(
        dataframe=test_df,
        x_col="image_path",
        y_col="dx",
        target_size=(IMG_SIZE, IMG_SIZE),
        batch_size=BATCH_SIZE,
        class_mode="categorical",
        shuffle=False  # Important to keep order for evaluation
    )
else:
    print(" Warning: test_df is not defined or empty!")

```

**Optimizing the CNN model**

```python
from tensorflow.keras.applications import MobileNetV2
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout, GlobalAveragePooling2D
from tensorflow.keras.optimizers import Adam

# Define the number of classes
NUM_CLASSES = train_df["dx"].nunique()  # Count unique labels

# Load the MobileNetV2 model without the top layer
base_model = MobileNetV2(weights="imagenet", include_top=False, input_shape=(IMG_SIZE, IMG_SIZE, 3))

# Freeze the pre-trained layers to retain learned features
base_model.trainable = False

# Define the new model
model = Sequential([
    base_model,
    GlobalAveragePooling2D(),
    Dense(128, activation='relu'),
    Dropout(0.5),
    Dense(NUM_CLASSES, activation='softmax')  # Output layer
])

# Compile the model
model.compile(optimizer=Adam(learning_rate=0.0001), loss='categorical_crossentropy', metrics=['accuracy'])

# Print model summary
model.summary()
```

TRAIN THE MODEL :

in small batches to prevent crashes.

```python
from tensorflow.keras.callbacks import EarlyStopping, ModelCheckpoint

EPOCHS = 15
## callbacks defined
early_stopping = EarlyStopping(monitor='val_loss', patience=3, restore_best_weights=True)
model_checkpoint = ModelCheckpoint("best_model.keras", save_best_only=True, monitor='val_loss')

# Train the model
history = model.fit(
    train_generator,
    validation_data=val_generator,
    epochs=EPOCHS,  # Increase to 15-20
    callbacks=[early_stopping, model_checkpoint]
)
```

```python
test_loss, test_accuracy = model.evaluate(test_generator)
print(f"Test Accuracy: {test_accuracy * 100:.2f}%")

```

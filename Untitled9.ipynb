```python
import zipfile
import os

# Define paths
zip_path = "/content/skin-cancer-mnist-ham10000.zip"
extract_path = "/content/skin_cancer_data"

# Extract the zip file
with zipfile.ZipFile(zip_path, 'r') as zip_ref:
    zip_ref.extractall(extract_path)

print("Dataset extracted successfully!")
```

```python
# List all files in the extracted folder
os.listdir(extract_path)
```

Load MetaData CSV

```python
import pandas as pd

# Load metadata
metadata_path = os.path.join(extract_path, "HAM10000_metadata.csv")
df = pd.read_csv(metadata_path)

# Show dataset info
print(df.info())
df.head()
```

Verifing Image Files

```python
import os

# Define image folder paths
image_folder_1 = "/content/skin_cancer_data/HAM10000_images_part_1/"
image_folder_2 = "/content/skin_cancer_data/HAM10000_images_part_2/"

# List some image files
print("Sample images from Part 1:", os.listdir(image_folder_1)[:5])
print("Sample images from Part 2:", os.listdir(image_folder_2)[:5])
```

Images Match MetaData

```python
all_images = set(os.listdir(image_folder_1) + os.listdir(image_folder_2))


missing_images = df[~df['image_id'].apply(lambda x: x + ".jpg").isin(all_images)]
print("Missing Images Count:", missing_images.shape[0])
```

Displaying Sample Images

```python
import matplotlib.pyplot as plt
import cv2

def show_images(image_ids):
    fig, axes = plt.subplots(1, len(image_ids), figsize=(15, 5))
    for idx, image_id in enumerate(image_ids):
        image_path = (
            os.path.join(image_folder_1, image_id + ".jpg")
            if os.path.exists(os.path.join(image_folder_1, image_id + ".jpg"))
            else os.path.join(image_folder_2, image_id + ".jpg")
        )
        image = cv2.imread(image_path)
        image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

        axes[idx].imshow(image)
        axes[idx].set_title(df[df["image_id"] == image_id]["dx"].values[0])
        axes[idx].axis("off")


sample_images = df.sample(5)["image_id"].tolist()
show_images(sample_images)
```
